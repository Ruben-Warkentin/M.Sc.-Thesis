This script finds charged residues based on their distance that experience electrostatic attraction. In proteins this would be considered a salt bridge, but here we are looking at these interactions between a protein (LoopDARPin; 4k5c) and a tetrasaccharide (SLeX).

```python
sb = pd.read_csv('09_ankyrin_hb_1.csv')
```

```python
#load csv file
df_wide = sb.pivot_table( index='Z', columns='Theta', values='Hbonds')

hm = sns.heatmap(df_wide, cmap="YlGnBu", linewidths=.003, cbar_kws={'label': '# of Hbonds'})
hm.invert_yaxis()
hm_fig = hm.get_figure() #this line may be an issue for the cluster, since there are no graphics
hm_fig.savefig("sb_hm.png")

#add a defined border between cells to make it more legible 
#flip the z axis
```
